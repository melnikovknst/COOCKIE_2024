```python
def conv_block(tensor, filters, kernel_size, activation, stft, strides=1, dropout=0.1):
  if stft:
    conv_layer = tf.keras.layers.Conv2D(filters=filters,
                                      kernel_size=kernel_size,
                                      strides=strides,
                                      padding='same',
                                      activation=None)(tensor)
  else:
    conv_layer = tf.keras.layers.Conv1D(filters=filters,
                                        kernel_size=kernel_size,
                                        strides=strides,
                                        padding='same',
                                        activation=None)(tensor)

  batch_norm_layer = tf.keras.layers.BatchNormalization()(conv_layer)
  activation_layer = tf.keras.layers.Activation(activation)(batch_norm_layer)
  dropout_layer = tf.keras.layers.Dropout(dropout)(activation_layer)

  if stft:
    pooling_layer = tf.keras.layers.MaxPool2D(pool_size=2,
                                            strides=2)(dropout_layer)
  else:
    pooling_layer = tf.keras.layers.MaxPool1D(pool_size=2,
                                              strides=2)(dropout_layer)

  return pooling_layer

def dense_block(tensor, units, activation, dropout=0.1):
  dense_layer = tf.keras.layers.Dense(units=units,
                                      activation=None)(tensor)
  batch_norm_layer = tf.keras.layers.BatchNormalization()(dense_layer)
  activation_layer = tf.keras.layers.Activation(activation)(batch_norm_layer)
  dropout_layer = tf.keras.layers.Dropout(dropout)(activation_layer)
  return dropout_layer

def get_model1(input_shape=(128000, 4), depth=8):
  input_layer = tf.keras.layers.Input(shape=input_shape)

  #Encoder
  for i in range(depth):
    if i == 0:
      conv = conv_block(tensor=input_layer,
                        filters=32,
                        kernel_size=3,
                        activation='relu',
                        stft=False,
                        strides=1,
                        dropout=0.1)
    else:
      conv = conv_block(tensor=conv,
                        filters=32,
                        kernel_size=3,
                        activation='relu',
                        stft=False,
                        strides=1,
                        dropout=0.1)

  #Fully-connected
  flatten_layer = tf.keras.layers.Flatten()(conv)

  dense = dense_block(tensor=flatten_layer,
                      units=100,
                      activation='relu')

  dense = dense_block(tensor=flatten_layer,
                      units=50,
                      activation='relu')

  output_layer = tf.keras.layers.Dense(units=2,
                                      activation='relu')(dense)

  model = tf.keras.models.Model(input_layer, output_layer)

  model.compile(optimizer='adam', loss='mse', metrics=['mape'])

  return model



def get_model2(input_shape=(8000, 32, 257), depth=8):
  input_layer = tf.keras.layers.Input(shape=input_shape)

  #Encoder
  for i in range(depth):
    if i == 0:
      conv = conv_block(tensor=input_layer,
                        filters=32,
                        kernel_size=3,
                        activation='relu',
                        strides=1,
                        dropout=0.1,
                        stft=True)
    else:
      conv = conv_block(tensor=conv,
                        filters=32,
                        kernel_size=3,
                        activation='relu',
                        strides=1,
                        dropout=0.1,
                        stft=True)

  #Fully-connected
  flatten_layer = tf.keras.layers.Flatten()(conv)

  dense = dense_block(tensor=flatten_layer,
                      units=100,
                      activation='relu')

  dense = dense_block(tensor=flatten_layer,
                      units=50,
                      activation='relu')

  output_layer = tf.keras.layers.Dense(units=2,
                                      activation='relu')(dense)

  model = tf.keras.models.Model(input_layer, output_layer)

  model.compile(optimizer='adam', loss='mse', metrics=['mape'])

  return model
```

```python
from datetime import datetime
import os
import wave
import numpy as np
from datetime import datetime, timedelta
import pandas as pd


def extract_datetime_from_filename(filename: str) -> datetime:
    # Извлекаем дату и время начала, игнорируя часть "to" и вторую временную метку
    date_str = filename.split('__')[1]  # Вторая часть строки содержит дату и время начала
    time_str = date_str.split('_')
    #print(time_str)
    # Форматируем строку как "YYYY-MM-DD HH:MM:SS"
    formatted_time_str = f"{time_str[4]}:{time_str[5]}:{time_str[6]} {time_str[0]}:{time_str[1]}:{time_str[2]}"

    # Преобразуем в datetime объект
    date_time = datetime.strptime(formatted_time_str, "%Y:%m:%d %H:%M:%S")

    return date_time
```

```python


def split_wav_file_by_channels(input_filename: str, output_folder: str, chunk_duration: int = 1):
    # Открываем исходный WAV файл
    with wave.open(input_filename, 'rb') as input_wav:
        # Получаем параметры WAV файла
        nchannels = input_wav.getnchannels()  # Количество каналов
        sampwidth = input_wav.getsampwidth()  # Ширина сэмпла (в байтах)
        framerate = input_wav.getframerate()  # Частота дискретизации
        nframes = input_wav.getnframes()  # Общее количество кадров

        total_samples_per_second = framerate * nchannels

        print(f"Число сэмплов в 1 секунде для  каналов: {framerate}")

        # Вычисляем количество фреймов в каждом куске (для 1 секунды)
        frames_per_chunk = framerate * chunk_duration

        # Создаем папку для выходных файлов, если ее нет
        if not os.path.exists(output_folder):
            os.makedirs(output_folder)

        # Читаем все данные из файла
        data = input_wav.readframes(nframes)

        # Получаем начальное время из имени файла
        start_time = extract_datetime_from_filename(input_filename)

        # Разбиваем данные на чанки по 1 секунде
        for second in range(0, nframes // framerate):
            chunk_start = second * frames_per_chunk * sampwidth*16
            chunk_end = chunk_start + frames_per_chunk * sampwidth*16
            chunk_data = data[chunk_start:chunk_end]

            # Обновляем время для текущей секунды
            chunk_start_time = start_time + timedelta(seconds=second)
            chunk_end_time = chunk_start_time + timedelta(seconds=chunk_duration)


            output_datetime = pd.to_datetime(f'{chunk_start_time.year}-0{chunk_start_time.month}-0{chunk_start_time.day} {chunk_start_time.hour}:{chunk_start_time.minute}:{chunk_start_time.second}', format = "%Y-%m-%d %H:%M:%S")

            # Создаем имя файла для сохранения данных
            output_filename = os.path.join(output_folder, f"{output_datetime}.npy")




            # Преобразуем данные в numpy массив, разделяя по каналам
            chunk_samples = np.frombuffer(chunk_data, dtype=np.int16)


           # Разделяем на каналы и записываем в выходной файл
            with open(output_filename, 'wb') as output_file:
                for channel in range(nchannels):
                    # Извлекаем данные для одного канала
                    channel_data = chunk_samples[channel::nchannels]

                    # Записываем данные в файл
                    np.save(output_file, channel_data)

            print(f"Сохранен файл: {output_filename}")

# Пример использования:
input_folder = '/content/drive/MyDrive/ML/COOKIE/info/data/act_dataset/данные_вибрации/2024_06_01/'  # Папка с исходными WAV файлами
output_folder = '/content/drive/MyDrive/ML/COOKIE/info/data/act_dataset/данные_вибрации/2024_06_01/splited'  # Папка для сохранения разбитых файлов

# Проходим по всем WAV файлам в папке `src` и разбиваем их по каналам
for filename in os.listdir(input_folder):
    if filename.endswith('.wav'):
        input_filepath = os.path.join(input_folder, filename)
        split_wav_file_by_channels(input_filepath, output_folder)
```

```python
import numpy as np
import librosa
import os
import matplotlib.pyplot as plt
import librosa.display

def apply_stft_to_numpy_file(file_path, n_fft=512, hop_length=256):
    # Загружаем numpy файл (предположим, что это одномерный массив)
    data = np.load(file_path)

    # Выводим форму данных для диагностики
    print(f"Форма загруженных данных: {data.shape}")

    # Преобразуем данные в тип с плавающей запятой (например, float32)
    data = data.astype(np.float32)

    # Получаем количество каналов и длину данных в одном канале
    num_channels = 4  # У нас 4 канала
    num_samples = len(data) // num_channels  # Количество семплов в одном канале

    if len(data) % num_channels != 0:
        print("Данные не делятся на равные части для 4 каналов.")
        return

    # Разбиваем данные на 4 канала
    data_split = np.split(data, num_channels)

    # Проверим, что данные разбились корректно
    print(f"Данные после разбиения: {[channel.shape for channel in data_split]}")

    # Создаем пустой массив для хранения результатов STFT
    stft_data = []

    # Применяем STFT для каждого канала
    for channel_idx in range(num_channels):
        channel_data = data_split[channel_idx]

        # Применяем STFT для канала (используем librosa для правильной работы с окнами и сдвигами)
        stft_result = librosa.stft(channel_data, n_fft=n_fft, hop_length=hop_length)

        # Получаем амплитуду (модуль спектра) для каждого канала
        stft_data.append(np.abs(stft_result))  # Добавляем результат для канала

    # Конвертируем список в numpy массив
    stft_data = np.array(stft_data)

    # Получаем имя файла без расширения
    base_name = os.path.splitext(os.path.basename(file_path))[0]

    # Формируем путь для сохранения нового numpy файла в нужное расположение
    output_path = f"/content/drive/MyDrive/Colab Notebooks/{base_name}_stft.npy"

    # Сохраняем данные в файл .npy
    np.save(output_path, {
        'num_samples': num_samples,
        'num_time_frames': stft_data.shape[2],  # число временных фреймов
        'num_frequency_bins': stft_data.shape[1],  # число частотных бинов
        'stft_data': stft_data
    })

    print(f"Данные сохранены в файл: {output_path}")

# Пример использования:
file_path = '/content/drive/MyDrive/Colab Notebooks/2024_6_1 17_12_29.npy'  # Замените на путь к вашему numpy файлу

apply_stft_to_numpy_file(file_path)
```
