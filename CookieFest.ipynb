```python
import pandas as pd

def get_outliers_idx(df, feature):
    # Проверяем, что столбец числовой
    if not pd.api.types.is_numeric_dtype(df[feature]):
        return pd.Series(False, index=df.index)  # Возвращаем пустую маску, если столбец не числовой

    # Вычисление IQR
    IQR = df[feature].quantile(0.75) - df[feature].quantile(0.25)
    low = df[feature].quantile(0.25) - (IQR * 1.5)
    up = df[feature].quantile(0.75) + (IQR * 1.5)

    # Определяем индексы выбросов
    outliers_index = (df[feature] < low) | (df[feature] > up)

    return outliers_index


def remove_outliers(df):
    # Сначала создаём булеву маску, которая будет True для строк, где есть выбросы в любом столбце
    all_outliers = pd.Series(False, index=df.index)  # Инициализируем маску с False

    for feature in df.columns:  # Проходим по всем столбцам
        # Получаем индексы выбросов для текущей фичи
        outliers_idx = get_outliers_idx(df, feature)
        print()

        # Обновляем маску выбросов, ставя True для тех строк, где есть выброс
        all_outliers = all_outliers | outliers_idx

    # Удаляем строки с выбросами
    df_cleaned = df[~all_outliers]  # Инвертируем маску и фильтруем DataFrame

    return df_cleaned

# Пример использования
stand1_1_df_cleaned = remove_outliers(stand1_1_df)
```
