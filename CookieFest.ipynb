```python
import pandas as pd


stand1_1_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data2024-06-01.csv', index_col = 0)
stand1_2_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data2024-08-23.csv', index_col = 0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df_1_1 = stand1_1_df.select_dtypes(include=['float64', 'int64'])

corr_df_1_1 = numeric_df_1_1.corr()

# Устанавливаем размер графика
plt.figure(figsize=(14, 12))  # Увеличиваем размер графика

# Отображаем тепловую карту
sns.heatmap(corr_df_1_1, annot=True, cmap='coolwarm', fmt='.2f', cbar=True,
            annot_kws={'size': 12})  # Увеличиваем размер шрифта чисел на карте

# Устанавливаем размер шрифта для подписей осей
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Устанавливаем размер шрифта для заголовка
plt.title('Матрица корреляции для stand_1_1', fontsize=16)

# Показываем график
plt.show()
```
